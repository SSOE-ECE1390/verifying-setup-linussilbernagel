```python
# Imports
import cv2

import os
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Load in image
# Read an image in using openCV
apollo_image = cv2.imread(os.path.relpath('Data/Apollo_11_Launch.jpg'), cv2.IMREAD_GRAYSCALE)

# View image
plt.imshow(apollo_image, cmap='gray')
plt.title("Apollo 11 Launch")
plt.show()
```

```python
# Crop image
cropped_apollo = apollo_image[50:600, 400:800]

# View image
plt.imshow(cropped_apollo, cmap='gray')
plt.title("Cropped Apollo 11 Launch")
plt.show()
```

```python
# Edit brightness
matrix = np.ones(cropped_apollo.shape, dtype="uint8") * 80

# Make it brighter
apollo_brighter = cv2.add(cropped_apollo, matrix)

# View image
plt.imshow(apollo_brighter, cmap='gray')
plt.title("Brighter Apollo 11 Launch")
plt.show()
```

```python
# Rotate image
apollo_rotate = cv2.rotate(apollo_brighter, cv2.ROTATE_90_CLOCKWISE)

# View image
plt.imshow(apollo_rotate, cmap='gray')
plt.title("Rotate Apollo 11 Launch")
plt.show()
```

```python
# Save modified image
cv2.imwrite("apollo_modified.png", apollo_rotate)
```
